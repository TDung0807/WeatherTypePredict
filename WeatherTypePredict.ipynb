# Library

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
```

# Prepare Dataset

```python
data = pd.read_csv("weather_classification_data.csv")
# show cloud type
cloud_cover = data["Cloud Cover"].unique()
print(cloud_cover)
# replace cloud type
cloud_mapping = {value: idx + 1 for idx, value in enumerate(cloud_cover)}
data["Cloud Cover"] = data["Cloud Cover"].replace(cloud_mapping)

# Show season type
season = data["Season"].unique()
print(season)
# replace season type
season_mapping={value: idx + 1 for idx, value in enumerate(season)}
data["Season"] = data["Season"].replace(season_mapping)

#show location type
location = data["Location"].unique()
print(location)
location_mapping={value: idx + 1 for idx, value in enumerate(location)}
data["Location"] = data["Location"].replace(location_mapping)

# Show weather type
weather = data["Weather Type"].unique()
print(weather)
# replace weather type
weather_mapping={value: idx + 1 for idx, value in enumerate(weather)}
data["Weather Type"] = data["Weather Type"].replace(weather_mapping)
```

```python
print(cloud_mapping)
print(season_mapping)
print(location_mapping)
print(weather_mapping)
```

```python
#Split data
X_train, X_test, y_train, y_test = train_test_split(data, data["Weather Type"], test_size=0.8)
```

# Create + training

```python

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", report)
```
